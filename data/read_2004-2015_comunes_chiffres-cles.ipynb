```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import show

%matplotlib inline
sns.set_context('notebook')
pd.options.mode.chained_assignment = None  # default='warn'
nan = np.nan # store numpy.nan in 'nan'
```

```python
data = pd.read_excel('2004-2015_comunes_chiffres-cles.xls', index_col=1) # , index_col=0, header=4)
# drop null rows & cols
data.dropna(axis=0,how='all', inplace=True)
data.dropna(axis=1,how='all', inplace=True)
# drop last 9 lines because they dont belong to the data
data.drop(data.tail(9).index, inplace=True)

data.head(10)
```

```python
#create the new headers (merge the category, header and year into one header)
category = ""
for col_index in range(0, len(data.columns)):
    cat_cell = data.iat[0, col_index]
    if(not pd.isnull(cat_cell)):
        category = "{}: ".format(cat_cell)
    col_name=data.iat[1, col_index]
    year = data.iat[2, col_index]
    if(pd.isnull(year)):
        year = ""
    else:
        year = " ({})".format(year)
    
    data.iat[1, col_index] = "{category}{col_name}{year}".format(category=category, col_name=col_name, year=year)
    # print("cat_cell: ", cat_cell, "  category: ", category, "  col_name: ", col_name, "  year: ", year)

#change the header row
data.columns = data.iloc[1]
#delete the row containing the headers
data = data[data.index != "Commune"]
#remove other superfluous rows
data = data[~pd.isnull(data.index)]

data.head(10)
```
