# Extracted each votation

1) Keep information about the votations

```python
def extractVotationByYearForCH(votations, date):
    votations2013 = votations[(votations.date >= np.datetime64(str(date))) & (votations.date < np.datetime64(str(date + 1)))]
    resultCH = votations2013[votations2013.id == 'CH']
    resultCH = resultCH.groupby(['idVotation', 'votation', 'id','municipality','date','label']).sum()
    resultCH = resultCH.unstack()

    #resultCH.set_index(['idVotation'])
    resultCH = resultCH.reset_index()
    resultCH.set_index('idVotation')
    resultCH.columns= ['idVotation', 'votation', 'id','municipality', 'date', 'bulletins','valables', 'inscrit', 'non','oui','percentage_oui','participation']
    columnsInt = ['idVotation', 'bulletins','valables', 'inscrit', 'non','oui']
    resultCH.head()
    for i in columnsInt:
        resultCH[i] = resultCH[i].astype(int)

    resultCH = resultCH.round(2) 
    resultCH = resultCH.set_index('idVotation')
    resultCH.to_csv('votations/'+str(date)+'/resultVotation.csv')
    return resultCH.index.unique()
```

2) Create a file for each votation

```python
def computeFileForOneVotation(df, id_vot,date):
    df = df[df.idVotation == id_vot]
    df = df.drop(['votation'], axis = 1)
    df = df.groupby(['idVotation', 'id','municipality','date','label']).sum()
    df = df.unstack()
    df = df.reset_index()
    df.columns= ['idVotation', 'id','municipality', 'date', 'bulletins','valables', 'inscrit', 'non','oui','percentage_oui','participation']
    columnsInt = ['idVotation', 'bulletins','valables', 'inscrit', 'non','oui']
    for i in columnsInt:
        df[i] = pd.to_numeric(df[i].astype(float)).map('{:.0f}'.format)

    for i in ['percentage_oui', 'participation']:    
        df[i] = df[i].astype('float')
        df[i] = df[i].round(2) 
        
    #Remove leading 0 in id, we cannot just convert to ID to int because we have such as 'ZH', 'FR',...
    df.id = df.id.astype(str).str.lstrip('0')
    df = df.set_index('id')
    #remove punctuation in munucipality name
    df.municipality = df.municipality.str.strip().str.split(' ', 1).str[-1]
    
    #remove all municipalities that contains 'étranger' or 'Auslander' (foreigner)
    # those are the swiss living in foreign countries, for sake of simplicity we remove them. We encounter the problem
    # that some of them have the same id as some Lakes and it create problem in the visualisation. 
    df = df[~ (df.municipality.str.contains('étranger') | df.municipality.str.contains('Ausland'))]
    
#     #Remove data about canton and CH=>the last 27 lines
#     df = df.iloc[:-27]
#     df.index = pd.to_numeric(df.index)
    df = df[df.index != 'CH']
#     print(df[df.index == 'CH'])

    f = pd.read_csv("data_to_be_processed/fusion_mappings.csv")
    f = f[f.date.str.contains('2016')]
    f.drop(['date'], axis=1, inplace=True)
    t = df.reset_index()
    fu = pd.merge(f, t, left_on='post_name', right_on='municipality', how='right')
    fu['municipality'] = fu['pre_name'].fillna(fu['municipality'])
    fu['id'] = fu['pre_id'].fillna(fu['id'])
    fu['id'] = fu['id'].astype(int)
    fu.drop(['pre_id', 'pre_name', 'post_id', 'post_name'], axis=1, inplace=True)
    fu.index = fu.id
    fu.drop(['id'], axis=1, inplace=True)
    fu.to_csv('votations/'+str(date)+'/'+str(id_vot)+'.csv')
    return fu
    
#     df.to_csv('votations/'+str(date)+'/'+str(id_vot)+'.csv')
#     return df



```

```python
for date in [2013,2014,2015,2016]: #
# for date in [2013]: #
    print('Year',date)
    id_votations = extractVotationByYearForCH(votations, date)
    votationsYear = votations[(votations.date >= np.datetime64(str(date))) & (votations.date < np.datetime64(str(date + 1)))]
    
#     for id_vot in id_votations[:1]:
    for id_vot in id_votations:
        print('ID',id_vot)
        df = computeFileForOneVotation(votationsYear, id_vot, date)
```
